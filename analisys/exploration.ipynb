**Bibliotecas**


```python
import pandas as pd
```

Realizar a leitura do dataset Trade Robo

```python
url = '../data/trade_robo.csv'
try:
    df_trade = pd.read_csv(url, sep=';', encoding='Latin-1')
except Exception as e:
    print(f'Error... Problemas ao carregar os dados {e}')
```

**Analise Exploratoria dos Dados (EDA)**

- Fundação dos Dados

```python
df_trade.head(2).style.set_caption('Lista as primeiras 5 linhas do dataset')
```

```python
# Supondo que você já carregou seu CSV/JSON em um DataFrame chamado 'df'

# 1. Converter a coluna 'L/P' para um formato numérico apropriado (Float)
# É fundamental garantir que a coluna 'L/P' seja numérica para a comparação
# (Pode ser necessário limpar o 'R$' e substituir vírgulas por pontos se os dados vierem como string)

# Exemplo de limpeza se 'L/P' for uma string como 'R$ 0,13':
df_trade['L/P'] = df_trade['L/P'].str.replace('R$', '').str.replace(',', '.').astype(float)

# 2. Criar uma nova coluna para identificar trades vencedores (Acerto)
# 1 se L/P > 0 (Lucro), 0 caso contrário (Prejuízo ou nulo)
df_trade['Acerto'] = (df_trade['L/P'] > 0).astype(int)
```

```python
# Agrupar por 'Estratégia' e calcular a média da coluna 'Acerto'
# O resultado será a acertividade em formato decimal (ex: 0.53)
acertividade_por_estrategia = df_trade.groupby('Estratégia')['Acerto'].mean().reset_index()

# Renomear a coluna para clareza
acertividade_por_estrategia.rename(columns={'Acerto': 'Acertividade (%)'}, inplace=True)

# Opcional: Converter para porcentagem
acertividade_por_estrategia['Acertividade (%)'] = (acertividade_por_estrategia['Acertividade (%)'] * 100).round(2)
```

```python
# Contar o número de trades por estratégia
contagem_trades = df_trade.groupby('Estratégia').size().reset_index(name='Total de Trades')

# Juntar os resultados
analise_final = pd.merge(acertividade_por_estrategia, contagem_trades, on='Estratégia')

# Ordenar pela acertividade (opcional)
analise_final = analise_final.sort_values(by='Acertividade (%)', ascending=False)
```
